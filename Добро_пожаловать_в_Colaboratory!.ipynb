```
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из файла
data = pd.read_csv('Current.csv', header=None)

# Взятие первых 1280 значений
current_values = data.iloc[:1280, 0]

average_current = current_values.mean()



power_values = (current_values ** 2) * 10
plt.plot(power_values)
plt.xlabel('Индекс')
plt.ylabel('Мощность')
plt.title('График значений мощности')
plt.show()


average_power_from_data = power_values.mean()
average_power_from_formula = (average_current ** 2) * 10

print(f'Среднее значение мощности из данных: {average_power_from_data}')
print(f'Среднее значение мощности по формуле: {average_power_from_formula}')
```

```
import matplotlib.pyplot as plt

# Создание графика
plt.figure(figsize=(12, 6))
plt.plot(range(len(current_values)), current_values, color='b', marker='o', linestyle='-')

# Добавление заголовка и меток осей
plt.title('График временного ряда')
plt.xlabel('Индекс')
plt.ylabel('Сила тока')

# Отображение сетки на графике
plt.grid(True)

# Отображение графика
plt.show()
```
